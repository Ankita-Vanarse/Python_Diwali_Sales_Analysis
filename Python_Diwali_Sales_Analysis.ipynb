## Exploratory Data Analysis


#### Gender

```python
df.columns
```

```python
sns.countplot(x = 'Gender',data = df)
plt.show()
```

```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by="Amount", ascending=False)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by="Amount", ascending=False)
sns.barplot(x= 'Gender',y='Amount' ,data = sales_gen)
plt.show()
```

##### From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

### Merital Status

```python
ax = sns.countplot(data = df , x = 'Marital_Status')

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

### Occupation

```python
sns.set(rc={'figure.figsize' :(15,5)})
ax = sns.countplot(data =df,x='Occupation',hue='Occupation', dodge=False, palette=palette)
palette = sns.color_palette("husl", len(df['Occupation'].unique()))
ax.legend([], [], frameon=False)
for bars  in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
sales_state = df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x= 'Occupation', y= 'Amount',hue='Occupation', dodge=False, palette=palette)
plt.show()
```

##### From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector 

### Product Category

```python
sns.set(rc={'figure.figsize':(20,6)})
ax = sns.countplot(data = df, x= "Product_Category",hue='Product_Category', dodge=False, palette=palette)
palette = sns.color_palette("husl", len(df['Product_Category'].unique()))
ax.legend([], [], frameon=False)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
palette = sns.color_palette("husl", len(sales_state['Product_Category'].unique()))
ax.legend([], [], frameon=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x= 'Product_Category',y= 'Amount',hue='Product_Category', dodge=False, palette=palette)
plt.show()
```

##### From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x= 'Product_ID',y='Orders',hue='Product_ID', dodge=False, palette=palette)
plt.show()
```
